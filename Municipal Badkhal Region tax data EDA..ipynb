## >>> Calculate how much unapproved area have diffrent type and their pending amount in ULB portal data and ndc portal data . 

## 1. NDC_PORTAL_unapproved_Amount

```python
NDC_unapproved =round(NDC_PORTAL.loc[NDC_PORTAL['Property falling in Approved Area/ Un approved (47)'] == 'UNAPPROVED'
              ].groupby('Category (20) - NDC').agg({'District (1)':'count',
                                                           'Total (31)=(26)+(27)+(28)+(29)+(30) - NDC ':'sum'
                                                          })).sort_values(by='Total (31)=(26)+(27)+(28)+(29)+(30) - NDC '
                                                                          ,ascending = False).reset_index().rename(columns={
    'District (1)': 'Total_Unapproved_Houses'
})
```

```python

NDC_unapproved
```

```python
categories = ['Residential', 'Mix Use', 'Special Category', 'Commercial', 'Institutional', 'Industrial', 'Vacant Plot']
total_unapproved_houses = NDC_unapproved['Total (31)=(26)+(27)+(28)+(29)+(30) - NDC ']
categories = categories[:-2]
total_unapproved_houses = total_unapproved_houses[:-2]


fig, ax = plt.subplots()
plt.suptitle("---------------NDC_unapproved_Data-----------")
ax.pie(total_unapproved_houses, labels=categories, autopct='%1.1f%%', startangle=90,radius=1)

plt.show()
```

```python
# Define your data
categories = ['Residential', 'Mix Use', 'Special Category', 'Commercial', 'Institutional', 'Industrial', 'Vacant Plot']
total_unapproved_houses = NDC_unapproved['Total (31)=(26)+(27)+(28)+(29)+(30) - NDC ']

# Filter out "Industrial" and "Vacant Plot" categories
categories = categories[:-2]
total_unapproved_houses = total_unapproved_houses[:-2]

# Create donut chart
fig, ax = plt.subplots()
ax.pie(total_unapproved_houses, labels=categories, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.5, 'edgecolor': 'white'})

# Add circle with white background to center of chart
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(center_circle)
plt.suptitle("---------------ULB_unapproved_Data-----------")

# Set aspect ratio to be equal so that chart is circular
plt.axis('equal')

# Show chart
plt.show()
```
